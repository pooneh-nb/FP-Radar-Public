```python
print("Number of Nodes: ", len(G.nodes()))
print("Number of Edges: ", len(G.edges()))
```

```python
#Lovain methof
```

```python
partition = community_louvain.best_partition(G, resolution=0.75, weight='weight')
```

```python
num_of_partitions = len(set(partition.values()))
print(num_of_partitions)
```

```python
api_hash_fract = {}
for key, value in partition.items():
    api_hash_fract[key] = value
```

```python
louvain_clusters = {}
for i in range(num_of_partitions):
    clust = []
    for key, value in api_hash_fract.items():
        if value == i:
            clust.append(key)
    louvain_clusters[i] = clust
```

```python
for i in range(num_of_partitions):
    print(len(louvain_clusters[i]))
```

```python
with open("/home/c6/Desktop/OpenWPM/myCodes/AST/jsons/fingerprinting_js2.txt", 'r') as file:
    bench_lib = file.read().splitlines()


for key_group, value_api in louvain_clusters.items():
    danger = 0
    for api in value_api:
        if api in bench_lib:
            print(api)
            danger+=1
    print(key_group, 100*danger/(len(bench_lib)))
```

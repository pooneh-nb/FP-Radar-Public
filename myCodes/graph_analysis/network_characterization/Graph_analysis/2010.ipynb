```python
from networkx.readwrite import json_graph
import networkx as nx
import json
import matplotlib.pyplot as plt
import community as community_louvain
import math
import utilities
```

```python
with open("/home/c6/Desktop/OpenWPM/jsons/AST/CDX_results/Graphs/2010/Graph.json", 'rt') as gr:
    g = json.load(gr)
```

```python
G = json_graph.node_link_graph(g)
```

```python
print(type(G))
```

```python
# Print number of nodes and edges
```

```python
print("Number of Nodes: ", len(G.nodes()))
print("Number of Edges: ", len(G.edges()))
```

```python
#Degree and aggregated weight (wtf) calculation
```

```python
for node_temp in G.nodes():
    wt = 0
    for nbr, eattr in G.adj[node_temp].items():
        wt = wt + eattr['weight']
    G.nodes[node_temp]['node_wtf'] = wt
    G.nodes[node_temp]['deg'] = len(G.adj[node_temp])
```

```python
# wtf between edges
for edge in G.edges():
    G.edges[edge]['wtf'] = G.edges[edge]['weight'] * math.log2(3)
    print(G.edges[edge]['wtf'])
    
```

```python
# Fature Extraction
all_nodes = utilities.read_json("/home/c6/Desktop/OpenWPM/jsons/Prediction/2011/node_list.json")
print(len(all_nodes))
count = 0
edge_set = set()
for idx, key in enumerate(all_nodes):
    for i in range(idx + 1, len(all_nodes)):
        #print(all_nodes[idx], all_nodes[i])
        if G.has_edge(all_nodes[idx], all_nodes[i]):
        #print(all_nodes[idx], all_nodes[i])
            count += 1
print(count)
print(len(edge_set))
```

```python
partition = community_louvain.best_partition(G, resolution=0.80, weight='weight')
```

```python
# number of partinions
```

```python
num_of_partitions = len(set(partition.values()))
print(num_of_partitions)
```

```python
api_hash_fract = {}
for key, value in partition.items():
    api_hash_fract[key] = value
louvain_clusters = {}
for i in range(num_of_partitions):
    clust = []
    for key, value in api_hash_fract.items():
        if value == i:
            clust.append(key)
    louvain_clusters[i] = clust
for i in range(num_of_partitions):
    print(len(louvain_clusters[i]))
```

```python
with open("2010_Lovain_clusters.json", 'w') as l:
    json.dump(louvain_clusters,l, indent=4)

```

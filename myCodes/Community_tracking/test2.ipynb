```python
## tilepuzzle.py


import sys
import copy

def tilepuzzle(start,goal,depth):
    return reverse(statesearch([start],goal,[],depth))

def statesearch(unexplored,goal,path,depth):
    if unexplored == []:
        return []
    elif goal == head(unexplored):
        return cons(goal,path)
    elif len(path) > depth:
        return []
    else:       
        result = statesearch(generateNewStates(head(unexplored)),
                             goal,
                             cons(head(unexplored), path),
                             depth)
 
        if result != []:
            return result
        else:
            return statesearch(tail(unexplored),
                               goal,
                               path,
                               depth)


def findZero(currState):

    for i in range(3):
        for j in range(3):
            if currState[i][j] == 0:
                return (i,j)
            else:
                continue


def tileUp(currState):

    index = findZero(currState)
    i = index[0]
    j = index[1]

    if i == 0:
        return []
    else:
        tempState = copy.deepcopy(currState)
        tempState[i][j] = currState[i-1][j]
        tempState[i-1][j] = 0
        return [tempState]

def tileDown(currState):

    index = findZero(currState)
    i = index[0]
    j = index[1]

    if i == 2:
        return []
    else:
        tempState = copy.deepcopy(currState)
        tempState[i][j] = currState[i+1][j]
        tempState[i+1][j] = 0
        return [tempState]

def tileLeft(currState):

    index = findZero(currState)
    i = index[0]
    j = index[1]

    if j == 0:
        return []
    else:
        tempState = copy.deepcopy(currState)
        tempState[i][j] = currState[i][j-1]
        tempState[i][j-1] = 0
        return [tempState]

def tileRight(currState):

    index = findZero(currState)
    i = index[0]
    j = index[1]

    if j == 2:
        return []
    else:
        tempState = copy.deepcopy(currState)
        tempState[i][j] = currState[i][j+1]
        tempState[i][j+1] = 0
        return [tempState]

def generateNewStates(currState):
    tempList = (tileUp(currState)) + (tileLeft(currState)) + (tileDown(currState)) + (tileRight(currState))
    return tempList

def reverse(lst):
    return lst[::-1]

def head(lst):
    return lst[0]

def tail(lst):
    return lst[1:]

def cons(item,lst):
    return [item] + lst

## tilepuzzle([[1,2,3],[4,5,6],[7,8,0]],[[1,2,3],[4,5,6],[7,0,8]],10)
```

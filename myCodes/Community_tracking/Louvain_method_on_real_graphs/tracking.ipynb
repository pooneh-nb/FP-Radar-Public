```python
huge_cluster_2019 = 16
huge_cluster_2018 = 6
huge_cluster_2017 = 3
huge_cluster_2016 = 6
huge_cluster_2015 = 1
```

### Comparing 2018 with the flagged community in 2019

```python
similarity = set(flagged_cluster_2019).intersection(set(flagged_cluster_2018))
difference =  set(flagged_cluster_2019).difference(set(flagged_cluster_2018))
print("Similarity: ", len(similarity), " Difference: ", len(difference))
```

```python
clusters_2018_dir = utilities.get_files_in_a_directory("/home/c6/Desktop/Pouneh_clusters/2018")
dicc = []     

for file in clusters_2018_dir:
    cluster_i = file.split("/")[-1].split('.')[0]
    if int(cluster_i) != huge_cluster_2018:
        cluster = []
        with open(file, 'r') as f:
            for line in f.readlines():
                cluster.append(line.rstrip('\n'))
        #print("cluster ", i_u," length:", len(cluster))
        similarity = set(cluster).intersection(set(flagged_cluster_2019))
        difference =  set(cluster).difference(set(flagged_cluster_2019))
        dicc.append({"cluster_i": cluster_i, 
                    "length_cluster" : len(cluster), 
                     "similarity": len(similarity), 
                     "difference": len(difference)})

display(pd.DataFrame(dicc))
```

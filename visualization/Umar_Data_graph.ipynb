```python
# make_partitions
import community as community_louvian
import networkx as nx
import json


with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_node.json", "rt") as nod:
    nodes_list = json.load(nod)
nodes = [tuple(l) for l in nodes_list]
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_edge.json", "rt") as ed:
    edges_list = json.load(ed)
edges = [tuple(ee) for ee in edges_list]

#edge_weight = []
#for ed in edges_list:
    #edge_weight.append(edges_list[i][2])
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
# do partitioning
partition = community_louvian.best_partition(G, weight='weight', resolution=0.94)
num_of_partitions = len(set(partition.values()))
print(num_of_partitions)
```

```python
# do positioning
import utilities
pos = nx.spring_layout(G, weight='weight')
print(len(pos))
utilities.write_dill("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_pos.json", pos)
```

```python
# Assign new positions
import dill
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_api_hash_frac_louvian_0.94.json", 'rt') as node:
    nodes = json.load(node)
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_pos.json", 'rb') as data:
    position = dill.load(data)

for api_key, pos in position.items():
    nodes[api_key]['x_pos'] = pos.item(0)
    nodes[api_key]['y_pos'] = pos.item(1)

with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned_api_hash_frac_louvian_0.94.json", 'w') as pos:
    json.dump(nodes, pos, indent=4)
```

```python
#assign apis to groups
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/api_hash_frac.json", "rt") as api:
    api_hash_fract = json.load(api)
louvain_part = {}
# assign each api to a group which is made by louvain
for key, value in partition.items():
    api_hash_fract[key]['louvain_partitioning'] = value
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/api_hash_frac_louvian_0.94.json","w") as lo:
    json.dump(api_hash_fract, lo, indent=4)
```

```python
### start from here if you need resolution 0.95
from collections import OrderedDict
from operator import getitem

with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/api_hash_frac_louvian_0.94.json", "rt") as api:
    api_clustered = json.load(api)
louvian_clustered = {}
for api_key, value in api_clustered.items():
    louvian_clustered[api_key] = {'weight': value['node_weight'], 'script': value['scripts'], 'group': value['louvain_partitioning']}


sorted_dict = OrderedDict(sorted(louvian_clustered.items(), key = lambda x:getitem(x[1],'weight'), reverse=True))

with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/sorted_louvian_0.94.json","w") as lo:
    json.dump(sorted_dict, lo, indent=4)
```

```python
##Clustering based on group
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/api_hash_frac_louvian_0.94.json", "rt") as api:
    api_louvain = json.load(api)
louvain_group = {}
#print(num_of_partitions)
print(len(api_louvain.items()))
api_list  = []
fraction = 0
for i in range(num_of_partitions):
    for api_key, value in api_louvain.items():
        if value['louvain_partitioning'] == i:
            #print(api_key, value['louvain_partitioning'])
            api_list.append(api_key)
            fraction = fraction + value['node_weight']
    louvain_group[str(i)] = {'api_list': api_list, 'fraction': fraction}
    fraction = 0
    api_list  = []
    
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/grouped_dict_0.94.json","w") as lo:
    json.dump(louvain_group, lo, indent=4)
```

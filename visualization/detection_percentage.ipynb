```python
# Libraries
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

### Detection Yimes 

```python
subgroup_names=['In Time', 'Early', 'Late', '']
subgroup_size=[29,14,13,44]
group_names=['Publicly disclosed', 'Undisclosed']
group_size=[56,44]
 
# Create colors
colors=["#FFC166", "#FFEB3B", "#F77B72", "#82C785"]
dis, undis = ["#FFFAFA", "#000000"]
 
# First Ring (outside)
fig, ax = plt.subplots(figsize=(8, 6), dpi=80)
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.5, labels=group_names, colors=["#c0c0c0", "#82C785"]
                 )
plt.setp( mypie, width=0.4, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.2, labels=subgroup_names, labeldistance=0.7, colors=["#FFC166", "#FFEB3B", "#F77B72", "#FFFFFF"])
plt.setp( mypie2, width=0.2, edgecolor='white')
plt.margins(0,0)

plt.legend(mypie, group_names, loc='lower left', title="Detection")
ax.set_title('Detection Times', fontsize=17, pad=25)
plt.axis('equal')
plt.tight_layout()
# show it
plt.show()
```

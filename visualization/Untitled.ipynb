```python
# make_partitions
import community as community_louvian
import networkx as nx
import json


with open("/home/c6/Desktop/OpenWPM/jsons/temporal_graphs/node_tuple_merge_2019_2020.json") as nod:
    nodes_list = json.load(nod)
nodes = [tuple(l) for l in nodes_list]
print(nodes)
with open("/home/c6/Desktop/OpenWPM/jsons/temporal_graphs/edge_tuple_merge_2019_2020.json") as ed:
    edges_list = json.load(ed)
edges = [tuple(ee) for ee in edges_list]

#edge_weight = []
#for ed in edges_list:
    #edge_weight.append(edges_list[i][2])
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
# do partitioning
partition = community_louvian.best_partition(G, weight='weight', resolution=0.94)
num_of_partitions = len(set(partition.values()))
print(num_of_partitions)
```

```python
#assign apis to groups
# assign each api to a group which is made by louvain
import pandas as pd
df = pd.DataFrame(columns=['index', 'class'])
for key, value in partition.items():
    df['index'] = key
    df['class'] = value
print(df)
```

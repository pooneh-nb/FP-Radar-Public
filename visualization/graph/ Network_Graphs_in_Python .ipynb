```python
import plotly.graph_objects as go
import community as community_louvain
import networkx as nx
import json
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import utilities
```

```python
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_node.json", "rt") as nod:
    nodes_list = json.load(nod)
nodes = [tuple(l) for l in nodes_list]
with open("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_edge.json", "rt") as ed:
    edges_list = json.load(ed)
edges = [tuple(ee) for ee in edges_list]
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
partition = community_louvain.best_partition(G,  weight='weight', resolution=0.94)
```

```python
pos = nx.spring_layout(G, weight='weight')
print(type(pos))
print(pos)
utilities.write_dill("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_pos.json", pos)
```

```python
position = utilities.read_dill("/home/c6/Desktop/OpenWPM/jsons/AST/umar_ds/pruned/pruned_pos.json")
for i, j in position.items():
    print(j.item((0)))
```

```python
nodes = [tuple(l) for l in nodes_list]
node_list = []
print(len(nodes))
for key_api, value in nodes:
    if value['node_weight'] > 0.1:
        nodes.append((key_api, {'node_weight': value['node_weight']}))
print(len(node_list))
print(node_list)
```

```python
cmap = cm.get_cmap('viridis', max(partition.values()) + 1)
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=40,
                       cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

```python
nx.set_node_attributes(G, pos, 'coord')
```

```python
edge_x = []
edge_y = []
for edge in G.edges():
    print(edge)
    x0, y0 = G.nodes[edge[0]]['coord']
    x1, y1 = G.nodes[edge[1]]['coord']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)
    print(x0,y0)

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines')

node_x = []
node_y = []
for node in G.nodes():
    x, y = G.nodes[node]['coord']
    node_x.append(x)
    node_y.append(y)

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        # colorscale options
        #'Greys' | 'YlGnBu' | 'Greens' | 'YlOrRd' | 'Bluered' | 'RdBu' |
        #'Reds' | 'Blues' | 'Picnic' | 'Rainbow' | 'Portland' | 'Jet' |
        #'Hot' | 'Blackbody' | 'Earth' | 'Electric' | 'Viridis' |
        colorscale='Greens',
        reversescale=True,
        color='black',
        size=10,
        colorbar=dict(
            thickness=15,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        ),
        line_width=2))
```

```python
node_adjacencies = []
node_text = []
for node, adjacencies in enumerate(G.adjacency()):
    node_adjacencies.append(len(adjacencies[1]))
    node_text.append('# of connections: '+str(len(adjacencies[1])))

node_trace.marker.color = node_adjacencies
node_trace.text = node_text
```

```python
fig = go.Figure(data=[edge_trace, node_trace],
             layout=go.Layout(
                title='<br>Network graph made with Python',
                titlefont_size=16,
                showlegend=False,
                hovermode='closest',
                margin=dict(b=20,l=5,r=5,t=40),
                annotations=[ dict(
                    text="Python code: <a href='https://plotly.com/ipython-notebooks/network-graphs/'> https://plotly.com/ipython-notebooks/network-graphs/</a>",
                    showarrow=False,
                    xref="paper", yref="paper",
                    x=0.005, y=-0.002 ) ],
                xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )
fig.show()
```
